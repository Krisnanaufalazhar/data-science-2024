# Normalisasi String

## Normalisasi Kolom Kota

```
df['kota'] = df['kota'].str.strip().str.title()
```

```
df['kondisi'] = df['kondisi'].str.strip().str.lower()
```

# Menangani Missing Values

## Mengisi Missing Value luas_m2

```
df['luas_m2'].fillna(df['luas_m2'].median(), inplace=True)
```

## Mengisi Missing Value harga_juta

```
df['harga_juta'].fillna(df['harga_juta'].median(), inplace=True)
```

## Mengisi Missing Value kamar

```
df['kamar'].fillna(df['kamar'].median(), inplace=True)
```

## Mengisi Missing Value kategorik

```
df['kondisi'].fillna(df['kondisi'].mode()[0], inplace=True)
```

## Validasi Missing Values

```
df.isnull().sum()
```

# Menangani Outlier dengan IQR Fence

## Membuat Fungsi IQR

```
def remove_outlier_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - (1.5 * IQR)
    upper = Q3 + (1.5 * IQR)

    data = data[
        (data[column] >= lower) &
        (data[column] <= upper)
    ]

    return data
```

## Menghapus Outlier harga_juta

```
df = remove_outlier_iqr(df, 'harga_juta')
```

## Menghapus Outlier luas_m2

```
df = remove_outlier_iqr(df, 'luas_m2')
```

# Membersihkan Tahun Bangun Tidak Wajar

```
df = df[df['tahun_bangun'] >= 1950]
```
