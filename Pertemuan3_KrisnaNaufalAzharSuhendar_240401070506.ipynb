# Menangani Outlier dengan IQR Fence

## Membuat Fungsi IQR

```python
def remove_outlier_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - (1.5 * IQR)
    upper = Q3 + (1.5 * IQR)

    data = data[
        (data[column] >= lower) &
        (data[column] <= upper)
    ]

    return data
```

## Menghapus Outlier harga_juta

```python
df = remove_outlier_iqr(df, 'harga_juta')

print("Outlier harga_juta berhasil dihapus")
```

## Menghapus Outlier luas_m2

```python
df = remove_outlier_iqr(df, 'luas_m2')

print("Outlier luas_m2 berhasil dihapus")
```
